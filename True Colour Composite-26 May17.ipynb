# 1. Load packages

```r
library(raster)
library(rgdal)
library(gdalUtils)
```

# 2. Get paths

```r
files.B2 <- dir("/data/MTDA/CGS_S2_RADIOMETRY/2017/05/26/",pattern="B02",full.names=T,recursive = T)
files.B3 <- dir("/data/MTDA/CGS_S2_RADIOMETRY/2017/05/26/",pattern="B03",full.names=T,recursive = T)
files.B4 <- dir("/data/MTDA/CGS_S2_RADIOMETRY/2017/05/26/",pattern="B04",full.names=T,recursive = T)
head(files.B2)
```

# 3. Make spatial vrt per band (combining all tiles)

```r
gdalbuildvrt(files.B2,"B2.vrt")
gdalbuildvrt(files.B3,"B3.vrt")
gdalbuildvrt(files.B4,"B4.vrt")
```

# 4. Make multiband vrt

```r
gdalbuildvrt(c("B2.vrt","B3.vrt","B4.vrt"),"RGB.vrt",separate=T)
```

# 5. Plot

```r
im <- stack("RGB.vrt")
plotRGB(im,3,2,1,scale=10000,stretch="lin")
```
